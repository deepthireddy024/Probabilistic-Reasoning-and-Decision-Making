```python
import random
```

```python
import matplotlib.pyplot as plt
```

```python
def probability(Z,s,alpha):
    p = (1-alpha)/(1+alpha)
    p *= pow(alpha,abs(Z-s))
    return p
```

```python
def estmateProbability(N_iter,Z,n_bits,bit,bit_value,alpha):
    if(bit==0 or bit>n_bits):
        print("invalid bit number")
        return
    num = 0
    denom = 0
    for i in range(N_iter):
        s = random.randint(0, pow(2,n_bits)-1)
        bs = '{0:010b}'.format(s)
        bs_bit = int(bs[-1*bit:-1*bit+1])
        
        p = probability(Z,s,alpha)
        
        if(bs_bit==bit_value):
            num += p
        denom += p      
        
    return num/(denom+pow(10,-8))
```

```python
#Number of samples in each iteration
ls = [i for i in range(10,100,10)]+ [i for i in range(100,1000,100)] + [i for i in range(1000,10000,1000)] + [i for i in range(10000,100000,10000)]+[i for i in range(100000,1000000,100000)]+[i for i in range(1000000,21000000,1000000)] 
```

## P(B2 = 1 | Z = 128)

```python
x = []
y = []
for i in ls:
    n = i
    es = estmateProbability(n,128,10,2,1,0.1)
    x.append(i)
    y.append(es)
    if(n>15000000):
        print("Number of samples:" + str(n))
        print("Estimate of probability:"+str(es))
        print("************************************************")
```

```python
plt.plot(x,y)
plt.title("P(B2 = 1 | Z = 128)")
plt.show()
```
